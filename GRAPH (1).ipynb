```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('movies.csv')
```

```python
genre_counts = movies['genres'].value_counts()
```

```python
print(genre_counts)
```

### hard coding

```python
drama=movies.loc[movies['genres'].str.contains('Drama')].shape[0]
comedy=movies.loc[movies['genres'].str.contains('Comedy')].shape[0]
romance=movies.loc[movies['genres'].str.contains('Romance')].shape[0]
sfic=movies.loc[movies['genres'].str.contains('Science Fiction')].shape[0]
Hor=movies.loc[movies['genres'].str.contains('Horror')].shape[0]
Fam=movies.loc[movies['genres'].str.contains('Family')].shape[0]
Act=movies.loc[movies['genres'].str.contains('Action')].shape[0]
west=movies.loc[movies['genres'].str.contains('Western')].shape[0]
adv=movies.loc[movies['genres'].str.contains('Adventure')].shape[0]
ani=movies.loc[movies['genres'].str.contains('Animation')].shape[0]
thr=movies.loc[movies['genres'].str.contains('Thriller')].shape[0]
my=movies.loc[movies['genres'].str.contains('Mystery')].shape[0]
f=movies.loc[movies['genres'].str.contains('Fantasy')].shape[0]
crime=movies.loc[movies['genres'].str.contains('Crime')].shape[0]
history=movies.loc[movies['genres'].str.contains('History')].shape[0]
war=movies.loc[movies['genres'].str.contains('War')].shape[0]
music=movies.loc[movies['genres'].str.contains('Music')].shape[0]
docf=movies.loc[movies['genres'].str.contains('Documentary')].shape[0]
fo=movies.loc[movies['genres'].str.contains('Foreign')].shape[0]
tvm=movies.loc[movies['genres'].str.contains('TV Movie')].shape[0]
list=[drama,comedy,romance,sfic,Hor,Fam,Act,west,adv,ani,thr,my,f,crime,history,war,music,docf,fo,tvm]
names=['Drama','Comedy','Romance','Science Fiction','Horror','Family','Action','Western','Adventure','Animation','Thriller','Mystery','Fantasy','Crime','History','War','Music','Documentary','Foreign','TV Movie']
```

```python
plt.figure(figsize=(25,25),dpi=100)
plt.bar(names,list,color='#e8689b')
plt.title('Number of Movies per Genre',fontdict={'fontsize':30})
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

### using for

```python

genrelist=movies['genres'].str.findall(r'"name":\s?"(.*?)"')

names=[]
for i in range(len(genrelist)):
    for j in range(len(genrelist[i])):
        if genrelist[i][j] not in names:
            names.append(genrelist[i][j])



genre_counts=[movies['genres'].str.contains(genre).sum() for genre in names]

plt.figure(figsize=(14, 6))
plt.bar(names, genre_counts, color='skyblue')
plt.xticks(rotation=45)
plt.ylabel('Number of Movies')
plt.xlabel('Genre')
plt.title('Movie Count by Genre')

plt.show()
```
