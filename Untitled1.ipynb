```python
gas=pd.read_csv('gas_prices.csv')


plt.figure(figsize=(10,8),dpi=100)

plt.title('Gas prices (in USD)',fontdict={'fontname':'Constantia','fontsize':20})
#plt.plot(gas.Year,gas.Canada,'r.-',label='Canada')
#plt.plot(gas.Year,gas.USA,'b.-',label='USA',)
#plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea',)
#plt.plot(gas.Year,gas.Australia,'m.-',label='Australia')
#plt.plot(gas.Year,gas.Germany,'y*-',label='Germany')

for country in gas:
    if country!='Year':
        plt.plot(gas.Year,gas[country],marker='.',label=country)

plt.xlabel('YEAR')
plt.ylabel('US DOLLARS')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.legend()
plt.savefig('gas_prices.png',dpi=300)
plt.show()
```

### LOAD FIFA DATA

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

### HISTOGRAMS

```python
bins=np.arange(40,110,10)
plt.hist(fifa.Overall,bins=bins,color='#148c88')
plt.xticks(bins)
plt.ylabel('Number Of Players')
plt.xlabel('Skill Level')
plt.title('FIFA STATISTICS')
plt.show()
```

### PIE CHART

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].shape[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].shape[0]

A=['left','right']
B=['#f20f5a','#0ff2eb']

plt.pie([left,right],labels=A,colors=B,autopct='%f %%')
plt.title('Foot preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
fifa.Weight
```

```python
plt.style.use("ggplot")

light=fifa.loc[fifa.Weight<125].shape[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].shape[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight>175)].shape[0]
medium_heavy=fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].shape[0]
heavy=fifa.loc[fifa.Weight>=200].shape[0]

weights=[light,light_medium,medium,medium_heavy,heavy]
a=['Under 125','125-150','150-175','175-200','Over 200']

explode=[.4,0,0,0,.4]
plt.pie(weights,labels=a,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

### BOX PLOT

```python
plt.style.use('default')
plt.figure(figsize=(10,10),dpi=100)
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
Cambridge_United=fifa.loc[fifa.Club=='Cambridge United']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']
labels=['FC BARCELONA','REAL MADRID','CAMBRIDGE UNITED','NEW ENGLAND REVOLUTION']

boxes=plt.boxplot([barcelona,madrid,Cambridge_United,revs],tick_labels=labels,patch_artist=True)

for box in boxes['boxes']:
    box.set(color='r',linewidth=2)
    box.set(facecolor='#e6106d')
    

plt.show()
```
